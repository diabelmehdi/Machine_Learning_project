```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import numpy as np

%matplotlib notebook
```

```python
s1 = 'side1-0.5'
s2 = 'front-black0.5'
k1 = pd.read_csv('object1/'+s1+'.csv')
k2 = pd.read_csv('object2/'+s2+'.csv')
```

```python
k1.shape
```

```python
fig, (ax0, ax1) = plt.subplots(ncols=2, constrained_layout=True)

ax0.set_title('Object #1 '+s1)
ax0.plot(k1)
ax0.grid(True)

ax1.set_title('Object #2 '+s2)
ax1.plot(k2)
ax1.grid(True)

plt.show()
```

```python
i = 8000
c1 =k1.loc[i:]
c2 =k2.loc[i:]


yticks = np.arange(-800, 800, 200)
fig, (ax0, ax1) = plt.subplots(ncols=2, constrained_layout=True)



ax0.plot(c1)
ax0.set_title('Object #1 '+s1)
ax0.set_yticks(yticks)
ax0.grid(True)


ax1.plot(c2)
ax1.set_title('Object #2 '+s2)
ax1.set_yticks(yticks)
ax1.grid(True)


plt.show()
```
